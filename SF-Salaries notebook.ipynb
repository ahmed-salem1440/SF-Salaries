```python
#starting with dropping the missing data in 'BasePay' column 
df.drop(index = df[df['BasePay'] == 'Not Provided'].index,inplace = True)
```

```python
#then convert the data type of it to numeric
df['BasePay'] = pd.to_numeric(df['BasePay'])
```

```python
df['OvertimePay'] = pd.to_numeric(df['OvertimePay'])
```

```python
df['OtherPay'] = pd.to_numeric(df['OtherPay'])
```

```python
df['Benefits'] = pd.to_numeric(df['Benefits'])
```

**What was the average BasePay of all employees per years from 2011 to 2014?**

```python
years = [2011,2012,2013,2014]
for year in (years):
    avg = df[df['Year'] == year ]['BasePay'].mean()
    print('The Average base pay in {} is {} '.format(year,avg) )
```

```python
#or by using .grouby() method
df.groupby('Year').mean()['BasePay']
```

**How many people have the word Chief in thier job title?**

```python
(df['JobTitle']).str.contains('Chief' , case = False).sum()
```

```python
#using .apply() method to find 'Chief' in all cases
def findchief(x):
    if 'chief' in x.lower():
        return True
    else:
        return False
df['JobTitle'].apply(findchief).sum()
```
